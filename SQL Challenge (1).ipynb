#Solving the Challenge

#1. What are the names of the players whose salary is greater than 100,000?

Using PySpark

```
q1=players.filter(players['salary']>100000)
q1.show()
```

Using SQL Query

```
teams.createOrReplaceTempView("teams")
players.createOrReplaceTempView("players")
matches.createOrReplaceTempView("matches")
```

```
q1_sql=spark.sql("Select * from players Where salary>100000")
q1_sql.show()
```

#2. What is the team name of the player with player_id = 3?

Using PySpark

```
team_id=players.filter(players['player_id']==3).select(players['team_id'])
q2=teams.join(team_id,team_id.team_id==teams.team_id).select(teams['team_name'])
q2.show()
```

Using SQL Query

```
q2_sql=spark.sql("Select team_name from teams where team_id=(Select team_id from players where player_id=3)")
q2_sql.show()
```

#3. What is the total number of players in each team?

Using PySpark

```
q3=players.groupBy(players['team_id']).agg(count(players['player_id']).alias("Total_Players_in_Team")).select(players['team_id'],"Total_Players_in_Team")
q3.show()
```

Using SQL Query

```
q3_sql=spark.sql("Select team_id,Count(player_id) as Total_Players_in_Team from players group by team_id")
q3_sql.show()
```

#4. What is the team name and captain name of the team with team_id = 2?

Using PySpark

```
q4=teams.join(players,teams.captain_id==players.player_id).filter(teams['team_id']==3).select(teams['team_name'],players['player_name'])
q4.show()
```

Using SQL Query

```
q4_sql=spark.sql("Select teams.team_name,players.player_name from teams join players on teams.captain_id=players.player_id where teams.team_id=3 ")
q4_sql.show()
```

#5. What are the player names and their roles in the team with team_id = 1?

Using PySpark

```
q5=teams.join(players,players.team_id==teams.team_id).filter(teams['team_id']==1).select(players['player_name'],players['role'])
q5.show()
```

Using SQL Query

```
q5_sql=spark.sql("Select players.player_name,players.role from teams join players on teams.team_id=players.team_id where teams.team_id=1")
q5_sql.show()
```

#6. What are the team names and the number of matches they have won?

Using PySpark

```
sub=matches.groupBy(matches['winner_id'].alias("winner")).agg(count(matches['match_id']).alias('Count_of_Matches')).select('winner','Count_of_Matches')
q6=teams.join(sub,sub.winner==teams.team_id).select(teams.team_name,sub.Count_of_Matches)
q6.show()
```

Using SQL Query

```
q6_sql=spark.sql("Select teams.team_name,sub.Count_of_Matches from teams join (Select matches.winner_id as winner,Count(matches.match_id) as Count_of_Matches from matches group by matches.winner_id) as sub on teams.team_id=sub.winner")
q6_sql.show()
```

#7. What is the average salary of players in the teams with country 'USA'?

Using PySpark

```
sub=teams.filter(teams['country']=='USA').select(teams['team_id'])
q7=players.join(sub,sub.team_id==players.team_id).agg(avg(players['Salary']).alias('Average_Salary')).select('Average_Salary')
q7.show()
```

Using SQL Query

```
q7_sql=spark.sql("Select avg(salary) as Average_Salary from players join (Select team_id from teams where country='USA') as sub on players.team_id=sub.team_id ")
q7_sql.show()
```

#8. Which team won the most matches?

Using PySpark

```
sub=matches.groupBy(matches['winner_id'].alias("winner")).agg(count(matches['match_id']).alias('Count_of_Matches')).orderBy(desc('Count_of_Matches')).select('winner','Count_of_Matches').limit(1)
q8=teams.join(sub,sub.winner==teams.team_id).select(teams.team_name,sub.Count_of_Matches)
q8.show()
```

Using SQL Query

```
q8_sql=spark.sql("Select teams.team_name,sub.Count_of_Matches from teams join (Select matches.winner_id as winner,Count(matches.match_id) as Count_of_Matches from matches group by matches.winner_id order by Count_of_Matches desc limit 1) as sub on sub.winner=teams.team_id")
q8_sql.show()
```

#9. What are the team names and the number of players in each team whose salary is greater than 100,000?

Using PySpark

```
sub=players.filter(players['salary']>=100000).groupBy(players['team_id']).agg(count(players['player_id']).alias('Players_with_salary_gt_100000')).select(players['team_id'],'Players_with_salary_gt_100000')
q9=teams.join(sub,sub.team_id==teams.team_id).select(teams['team_name'],sub['Players_with_salary_gt_100000'])
q9.show()
```

Using SQL Query

```
q9_sql=spark.sql("Select teams.team_name,sub.Players_with_salary_gt_100000 from teams join (Select team_id,count(player_id) as Players_with_salary_gt_100000 from players where salary>=100000 group by team_id)as sub on sub.team_id=teams.team_id")
q9_sql.show()
```

#10. What is the date and the score of the match with match_id = 3?

Using PySpark

```
q10=matches.filter(matches['match_id']==3).select(matches['match_date'],matches['score_team1'],matches['score_team2'])
q10.show()
```

Using SQL Query

```
q10_sql=spark.sql("Select match_date,score_team1,score_team2 from matches where match_id=3")
q10_sql.show()
```
